```python
def get_chat_prompt(role, topic):
    
    # Define role behavior
    if role == "teacher":
        system_message = "You are a helpful teacher who explains concepts in simple terms."
        
    elif role == "interviewer":
        system_message = "You are an interviewer who asks challenging and relevant questions."
        
    elif role == "motivator":
        system_message = "You are a motivational speaker who encourages learning."
        
    else:
        system_message = "You are a helpful AI assistant."

    # Create ChatPromptTemplate
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    # Format messages
    messages = chat_template.format_messages(topic=topic)
    
    return messages


# Test
result = get_chat_prompt("teacher", "Neural Networks")

# Print result
for msg in result:
    print(f"{msg.type}: {msg.content}")
```

```python
def validate_inputs(audience, tone):
    
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]
    
    # Check audience
    if audience not in valid_audience:
        print("Invalid audience! Setting default → beginner")
        audience = "beginner"
    
    # Check tone
    if tone not in valid_tone:
        print("Invalid tone! Setting default → formal")
        tone = "formal"
    
    return audience, tone


# Test Cases
print(validate_inputs("beginner", "fun"))     # valid
print(validate_inputs("kids", "cool"))        # invalid
```

```python
def generate_prompt(topic, audience, tone, style):
    
    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)
    
    # Step 2: Select template based on style
    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} step by step for {audience} in a {tone} tone"
        )
        
    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 interview questions about {topic}"
        )
        
    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} as a story for {audience} in a {tone} tone"
        )
        
    else:
        return "Invalid style selected"
    
    # Step 3: Generate output
    if style == "interview":
        return template.format(topic=topic)
    else:
        return template.format(topic=topic, audience=audience, tone=tone)


# ✅ Test Cases
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print(generate_prompt("Python", "kids", "cool", "teaching"))   # invalid inputs test
print(generate_prompt("Machine Learning", "expert", "formal", "interview"))
```

```python
# One reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

# 5 different inputs
test_inputs = [
    ("AI", "beginner"),
    ("Python", "students"),
    ("Data Science", "analysts"),
    ("Machine Learning", "engineers"),
    ("Deep Learning", "researchers")
]

# Run loop
for topic, audience in test_inputs:
    result = reusable_template.format(topic=topic, audience=audience)
    print(result)
```
